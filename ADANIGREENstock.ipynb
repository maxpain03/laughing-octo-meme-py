# Simple Returns

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as dt
import matplotlib.pyplot as plt
```

```python
## Considering Time Frame as 1 year (19 Dec 2019 to 18 Dec 2020)
```

```python
adani = dt.DataReader ('ADANIGREEN.NS', data_source = 'yahoo', start = '2019-12-19')
adani
```

```python
# Now Finding Simple returns for Adani Green comapany's stock listed on NSE
```

```python
adani['simple_return'] = (adani['Adj Close'] / adani['Adj Close'].shift(1)) - 1
```

```python
adani['simple_return']
```

```python
adani['simple_return'].plot(figsize = (10, 5))
```

```python
# Average for total observations
```

```python
avg = round(adani['simple_return'].mean(),5)
avg
```

```python
# Stock market is active for 250 days in a year hence average should be multiplied by 250
```

```python
avg = avg*250
avg
```

```python
print (str(round(avg,5) * 100) + '%')
```

# Logarithmic Returns

```python
adani ['log_return'] = np.log(adani['Adj Close']/adani['Adj Close'].shift(1))

print (adani['log_return'])
```

```python
adani ['log_return'].plot(figsize = (8,5))
plt.show()
```

```python
# Now the daily average of Log returns is:
```

```python
adani_daily = round(adani['log_return'].mean(),5)
adani_daily
```

```python
# Converting this into 250 working days of a year
```

```python
adani_annual = round(adani_daily * 250, 5)
adani_annual
```

```python
print (str(round(adani_annual,5) * 100) + '%')
```
